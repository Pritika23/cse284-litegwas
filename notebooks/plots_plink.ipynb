```python
# Import libraries
%pylab inline
import os
import pandas as pd
from qqman import qqman
import matplotlib.pyplot as plt
```

```python
"""Question 2.1: Make Manhattan and QQ plot of the results"""
data = pd.read_csv("gwas_results.assoc.logistic", delim_whitespace=True)
fig, (ax0, ax1) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [2, 1]})
fig.set_size_inches((15, 5))
qqman.manhattan(data, ax=ax0)
qqman.qqplot(data, ax=ax1)
```
